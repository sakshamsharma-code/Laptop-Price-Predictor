```python
df.drop(columns=['weight','Number of Ratings', 'Number of Reviews'], axis=1,inplace=True)
```

```python
num=df.select_dtypes(exclude='object') 
```

```python
num.columns
```

```python
char=df.select_dtypes(include='object')  
```

```python
char.columns
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
char['brand'] = le.fit_transform(char['brand'])
char['processor_brand'] = le.fit_transform(char['processor_brand'])
char['processor_name'] = le.fit_transform(char['processor_name'])
char['processor_gnrtn'] = le.fit_transform(char['processor_gnrtn'])
char['ram_gb'] = le.fit_transform(char['ram_gb'])
char['ram_type'] = le.fit_transform(char['ram_type'])
char['ssd'] = le.fit_transform(char['ssd'])
char['hdd'] = le.fit_transform(char['hdd'])
char['os'] = le.fit_transform(char['os'])
char['os_bit'] = le.fit_transform(char['os_bit'])
char['graphic_card_gb'] = le.fit_transform(char['graphic_card_gb'])
char['warranty'] = le.fit_transform(char['warranty'])
char['Touchscreen'] = le.fit_transform(char['Touchscreen'].map({'Yes': 1, 'No': 0}))
char['msoffice'] = le.fit_transform(char['msoffice'].map({'Yes': 1, 'No': 0}))
char['rating'] = le.fit_transform(char['rating'])
```

```python
char
```

```python
clean_data=pd.concat([char,num],axis=1) 
```

```python
clean_data
```

```python
clean_data.to_csv(r"C:\Final Project\Laptop-Price-Predictor\data\Cleaned_Laptop_Price.csv", index=False)
```
